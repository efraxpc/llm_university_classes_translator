# C02 — Prompting

Técnicas de engenharia de prompts: zero-shot, few-shot, chain-of-thought e system prompts.

```python
import os
import anthropic
from dotenv import load_dotenv

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv("ANTHROPIC_API_KEY"))
MODEL = "claude-sonnet-4-6"

def chat(system: str, user: str, max_tokens: int = 512) -> str:
    resp = client.messages.create(
        model=MODEL,
        max_tokens=max_tokens,
        system=system,
        messages=[{"role": "user", "content": user}]
    )
    return resp.content[0].text
```

## 1. Zero-shot prompting

```python
resposta = chat(
    system="Você é um assistente especialista em inteligência artificial.",
    user="Classifique o sentimento: 'O produto chegou com defeito e o suporte não ajudou.'"
)
print("Zero-shot:", resposta)
```

## 2. Few-shot prompting

```python
few_shot_prompt = """
Classifique o sentimento como POSITIVO, NEGATIVO ou NEUTRO.

Exemplos:
Texto: "Adorei o produto, superou minhas expectativas!" → POSITIVO
Texto: "A entrega atrasou três dias." → NEGATIVO
Texto: "O produto chegou na data prevista." → NEUTRO

Texto: "O atendimento foi horrível e o produto veio quebrado." →
"""

resposta = chat(system="", user=few_shot_prompt)
print("Few-shot:", resposta)
```

## 3. Chain-of-Thought (CoT)

```python
cot_prompt = """
Resolva o problema passo a passo:

Uma loja tem 120 camisetas. Ela vende 35% na segunda-feira e 25% do restante na terça-feira.
Quantas camisetas restam após os dois dias?

Pense em voz alta antes de dar a resposta final.
"""

resposta = chat(system="Você é um assistente matemático.", user=cot_prompt)
print(resposta)
```

## 4. System prompt com persona

```python
system_persona = """
Você é DataBot, um assistente especializado em ciência de dados.
- Responda sempre de forma concisa e técnica.
- Use exemplos de código Python quando relevante.
- Se não souber algo, diga claramente.
"""

resposta = chat(
    system=system_persona,
    user="Como calcular a correlação de Pearson entre duas variáveis?"
)
print(resposta)
```

## 5. Prompt para saída estruturada (JSON)

```python
import json

json_prompt = """
Extraia as entidades do texto abaixo e retorne um JSON com as chaves:
"pessoas", "organizacoes", "locais"

Texto: "Elon Musk anunciou que a Tesla abrirá uma nova fábrica em Monterrey, México."

Retorne somente o JSON, sem explicações.
"""

resposta = chat(system="", user=json_prompt)
dados = json.loads(resposta)
print(json.dumps(dados, indent=2, ensure_ascii=False))
```
